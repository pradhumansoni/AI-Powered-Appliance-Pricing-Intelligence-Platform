### ***HTML Parsing and Data Extraction***

### Overview
This notebook parses raw HTML data collected from e-commerce product listings, extracts relevant product information, converts it into a structured pandas DataFrame, and exports the cleaned dataset as a CSV file for further analysis and modeling.

### Workflow
- Load raw HTML files
- Parse HTML using BeautifulSoup
- Extract product attributes using selectors
- Handle missing values and inconsistent fields
- Create a pandas DataFrame
- Export structured data to CSV

### Technologies Used
- Python
- BeautifulSoup4
- pandas
- NumPy

### Output
The final cleaned dataset is stored as a `.csv` file and can be used for:
- Exploratory Data Analysis (EDA)
- Machine Learning models
- Price prediction systems
- Market trend analysis

```python
# import importlib
# importlib.reload(s)
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import sys
sys.path.append("D:/Study/data_science/underpriced-listing-predictor/scraper")

import ac_selectors as s 
```

```python
# Open raw ac_full_dump file

with open('D:/Study/data_science/underpriced-listing-predictor/data/01.raw/refrigerator.html','r' , encoding='utf-8') as f:
    fridge_html = f.read()

soup = BeautifulSoup(fridge_html , 'lxml') # use beautiful soup class to check ac_html file
```

```python
containers = soup.find_all('div' , {'class':'sm-product has-tag has-features has-actions'}) # Containers containing each card data 

print(f"Total Cards -> {len(containers)}")
```

```python
s.FEATURES
```

```python
s.RATING
```

```python
# We see that The names of each appliance is inside "h2"
name = []
price = []
rating = [] 
ac_type = []
capacity = []
features = []



for i in containers:
    try:
        name.append(i.find(s.NAME['tag']).text)
    except:
        name.append(np.nan)

    try:
        price.append(i.find(s.PRICE['tag'] , s.PRICE['attrs']).text)
    except:
        price.append(np.nan)

    try:
        rating.append(i.find(s.RATING['container_tag'], s.RATING['container_attrs'])
                      .find(s.RATING['value_tag'] ,s.RATING['value_attrs'])
                      [s.RATING['style_tag']])
    except:
        rating.append(np.nan)


    x = i.find(s.FEATURES['container_tag'] , s.FEATURES['container_attrs']).find_all(s.FEATURES['item_tag'])


    try:
        ac_type.append(x[0].text)
    except:
        ac_type.append(np.nan)
    try:
        capacity.append(x[1].text)
    except:
        capacity.append(np.nan)
    
    sub_feature = []

    for j in range(2, 8):
        try:
            sub_feature.append(x[j].text)
        except:
            sub_feature.append(np.nan)
    features.append(sub_feature)
    
```

```python
import pandas as pd

df = pd.DataFrame({'name':name , 'price':price, 'rating': rating , 'ac_type':ac_type , 'capacity':capacity , 'features':features})

print(df.info())
```
