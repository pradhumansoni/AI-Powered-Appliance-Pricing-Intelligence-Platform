```python

# Price outliers
def detect_outliers_iqr(data):
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    return (data < Q1 - 1.5*IQR) | (data > Q3 + 1.5*IQR)

price_outliers = detect_outliers_iqr(df['price'])
print(f"Price outliers: {price_outliers.sum()} ({price_outliers.sum()/len(df)*100:.1f}%)")
print(f"Range: ₹{df[price_outliers]['price'].min():.0f} - ₹{df[price_outliers]['price'].max():.0f}")

# Visualize
fig, ax = plt.subplots(figsize=(10, 6))
ax.scatter(df[~price_outliers].index, df[~price_outliers]['price'], 
          alpha=0.5, label='Normal', s=30)
ax.scatter(df[price_outliers].index, df[price_outliers]['price'], 
          alpha=0.7, label='Outliers', s=50, color='red')
ax.set_title('Price Outliers Detection (IQR method)')
ax.set_xlabel('Product Index')
ax.set_ylabel('Price (₹)')
ax.legend()
plt.show()
plt.tight_layout()
```

```python
# For each category, deep dive
for cat in df['category'].unique():
    cat_df = df[df['category'] == cat]
    print(f"\n{'='*50}")
    print(f"{cat.upper()} - {len(cat_df)} products")
    print(f"{'='*50}")
    
    # Price statistics
    print(f"Price: mean={cat_df['price'].mean():.0f}, "
          f"median={cat_df['price'].median():.0f}, "
          f"std={cat_df['price'].std():.0f}")
    
    # Top features
    cat_features = [col for col in cat_df.columns if col.startswith(cat.lower()[:2])]
    print(f"\nTop 5 Features (by avg price difference):")
    
    feature_impact = {}
    for feat in cat_features[:10]:
        if cat_df[feat].notna().sum() > 0:
            with_feature = cat_df[cat_df[feat] == 1]['price'].mean()
            without_feature = cat_df[cat_df[feat] == 0]['price'].mean()
            feature_impact[feat] = with_feature - without_feature
    
    for feat, impact in sorted(feature_impact.items(), 
                               key=lambda x: abs(x[1]), 
                               reverse=True)[:5]:
        print(f"  {feat}: +₹{impact:.0f}")
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor as VIF

numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
# Select numeric features and convert to float64 to handle nullable types
X_numeric = df[numeric_cols].fillna(df[numeric_cols].median()).astype('float64')

# Calculate VIF
vif_data = pd.DataFrame()
vif_data['Feature'] = X_numeric.columns
vif_data['VIF'] = [VIF(X_numeric.values, i) for i in range(X_numeric.shape[1])]
vif_data = vif_data.sort_values('VIF', ascending=False)

print(vif_data)
# VIF > 10 indicates problematic multicollinearity
# VIF > 5 indicates moderate multicollinearity
```
